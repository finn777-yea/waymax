```python
import jax
import jax.numpy as jnp
import numpy as np
import mediapy
import dataclasses
import matplotlib.pyplot as plt

from waymax import config as _config
from waymax import dataloader
from waymax import datatypes
from waymax import visualization
from waymax import env as _env
from waymax import agents
from waymax import dynamics
```

```python
MAX_NUM_OBJECTS = 16

config = dataclasses.replace(_config.WOD_1_1_0_TRAINING, max_num_objects=MAX_NUM_OBJECTS)
data_iter = dataloader.simulator_state_generator(config)
scenario = next(data_iter)
```

```python
_config.ObjectType.VALID
```

```python
# TODO: check the scenario(state) for map informations

scenario.timestep

# plot the roadgraph
rg_points = scenario.roadgraph_points

where_valid = rg_points.valid

plt.scatter(
    x = rg_points.x[where_valid],
    y = rg_points.y[where_valid],
    s=0.1,
    c = rg_points.types[where_valid]
)
plt.show()
```

```python
# inspect the map element types
rg_points.types[where_valid]
```

```python
dynamics_model = dynamics.StateDynamics()

env_config = dataclasses.replace(
    _config.EnvironmentConfig(),
    max_num_objects=MAX_NUM_OBJECTS,
    controlled_object=_config.ObjectType.VALID
)

env = _env.MultiAgentEnvironment(
    dynamics_model,
    config=env_config)
```

```python
state = [env.reset(scenario)][0]

img = visualization.plot_simulator_state(state)
mediapy.show_image(img)
```
